```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "C:\\Users\\shane\\Downloads\\purchase_data.csv"
```

```python
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
total_users = len(purchase_data_df["SN"].unique())

total_users_df = pd.DataFrame({"Total Players": [total_users]})
total_users_df
```

```python
unique_items = len(purchase_data_df["Item Name"].unique())

average_price = purchase_data_df["Price"].mean()

total_revenue = purchase_data_df["Price"].sum()

total_purchases = purchase_data_df["Purchase ID"].count()
```

```python
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], 
                                      "Average Purchase Items": [average_price],
                                      "Total Number of Purchases": [total_purchases],
                                      "Total Revenue": [total_revenue]})

purchasing_analysis_df["Average Purchase Items"] = purchasing_analysis_df["Average Purchase Items"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)                                                                                                      
purchasing_analysis_df
```

```python
players = purchase_data_df.loc[:, ["SN", "Gender"]]

unique_players = players.drop_duplicates()

gender_count = unique_players["Gender"].value_counts()

gender_percent = (gender_count / total_users)*100

gender_demographics_df = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": gender_percent})

gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("%{:.2f}".format)
gender_demographics_df
```

```python
gp_list = purchase_data_df.loc[:, ["Gender", "Price"]]

gender_group = purchase_data_df.groupby("Gender")

gender_total = gp_list["Gender"].value_counts()

gender_total_purchase = gender_group["Price"].sum()

avg_purchase = gender_total_purchase / gender_total

avg_per_person = gender_total_purchase / gender_count

gender_analysis_df = pd.DataFrame({"Purchase Count": gender_total, 
                                   "Average Purchase Price": avg_purchase, 
                                   "Total Purchase Value": gender_total_purchase,
                                   "Avg Total Purchase per Person": avg_per_person})

gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map("${:.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:.2f}".format)
gender_analysis_df["Avg Total Purchase per Person"] = gender_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

gender_analysis_df
```

```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 99]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

ages = purchase_data_df.loc[:, ["SN", "Age","Price"]]

unique_age = ages.drop_duplicates()

unique_age["Age Ranges"] = pd.cut(unique_age["Age"], bins, labels=group_names)

players_age = unique_age.groupby("Age Ranges")

age_count = players_age["Age Ranges"].count()

age_percent = (age_count / total_users)*100

age_demographics_df = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

age_demographics_df
```

```python
ages["Age Ranges"] = pd.cut(ages["Age"], bins, labels=group_names, include_lowest=True)

age_purchase_count = players_age["Age Ranges"].count()

age_count_purchase = len(unique_age["SN"].unique())
 
age_purchase_total = players_age["Price"].sum()

age_purchase_avg = age_purchase_total / age_purchase_count

age_avg_person = age_purchase_total / age_count_purchase

age_purchase_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                               "Average Purchase Price": age_purchase_avg,
                               "Total Purchase Value": age_purchase_total})

age_purchase_df["Average Purchase Price"] = age_purchase_df["Average Purchase Price"].map("${:.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"].map("${:.2f}".format)

age_purchase_df
```

```python
spenders = purchase_data_df.loc[:, ["SN", "Price"]]

spender_group = spenders.groupby("SN")

purchase_count = spender_group["SN"].count()

total_spender = spender_group["Price"].sum()

avg_spender = total_spender / purchase_count

top_spender_df = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": avg_spender,
                              "Total Purchase Value": total_spender})

top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)

top_sorted_df = top_spender_df.sort_values(["Total Purchase Value"], ascending=False)
top_sorted_df.head()
```

```python
popular_items = purchase_data_df.loc[:, ["Item ID", "Item Name", "Price"]]

items_group = popular_items.groupby("Item ID")

item_count = items_group["Item Name"].value_counts()

popular_price = items_group["Price"]

popular_price = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"]

item_total = popular_price * item_count

most_popular_df = pd.DataFrame({"Purchase Count": item_count, 
                                "Item Price": popular_price, 
                                "Total Purchase Value": item_total})

most_popular_df["Item Price"] = most_popular_df["Item Price"].map("${:.2f}".format)
most_popular_df["Total Purchase Value"] = most_popular_df["Total Purchase Value"].map("${:.2f}".format)

most_popular_sorted = most_popular_df.sort_values(["Purchase Count"], ascending=False)

most_popular_sorted.head()
```
